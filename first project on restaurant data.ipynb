```python
# Data Analysis practice project for restraurant Data with pandas
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
# display the first rows of the dataset
tips.head()
```

```python
#how many weekdays are present in the dataset
tips.day.unique()
```

```python
#last 5 rows
tips.tail()
```

```python
#how many dinners and lunches data are there in the dataset
tips.time.value_counts()
```

```python
tips.time.value_counts().plot();
```

```python
tips.time.value_counts().plot(kind= 'bar');
```

```python
#what is the proportionof the dinner
tips.time.value_counts(normalize=True)
```

```python
#when do people pay more tips
## use groupby
tips.groupby(['time'])['tip'].mean()
```

```python
#lets plot the values
tips.groupby(['time'])['tip'].mean().plot.bar()
```

```python
#smoker/non-smoker, sex, size, day
tips.groupby(['smoker', 'sex'])['tip'].mean()
```

```python
tips.groupby(['smoker', 'sex'])['tip'].mean().unstack()
```

```python
#new column to calculate the percentage tip
tips['tip_percentage'] = tips.tip/tips.total_bill
```

```python
tips.head()
```

```python
#plot relationship between total bill and tips
tips.plot.scatter(x ='total_bill', y='tip');
```

```python
#visualizing using the seaborn library
sns.scatterplot(x='total_bill', y='tip', data = tips);
```

```python
#lets add the effect of the column sex to understand the data better
sns.scatterplot(x='total_bill', y='tip', data = tips, hue='sex');
```

```python
sns.scatterplot(x='total_bill', y='tip', data = tips, hue = 'time');
```

```python
sns.relplot(x='total_bill', y='tip', data = tips, col='sex', hue='time');
```

```python
sns.displot(x='total_bill', data = tips, col='time', kde=True);
```

```python
sns.pairplot(data=tips, hue='sex');
```
